```python
#import pandas
import pandas as pd

#set csv path
schools_csv = "/Users/katieboals/Downloads/schools_complete.csv"
students_csv = "/Users/katieboals/Downloads/students_complete.csv"


#read csv file
schools_data = pd.read_csv(schools_csv)
students_data = pd.read_csv(students_csv)
```

```python
#combined data with a little help from stack overflow, awesome tool
schools_complete = pd.merge(students_data, schools_data, how ="left", on=["school_name", "school_name"])
schools_complete.head()
```

```python
#find total number of unique schools

unique_schools = schools_complete["school_name"].unique()

total_schools = len(unique_schools)
total_schools
```

```python
#find the total students
total_students = schools_complete["student_name"].count()
total_students
```

```python
#find total budget
total_budget = schools_data["budget"].sum()
total_budget
```

```python
#find percentage of students who passed math

math_passing = schools_complete.loc[schools_complete["math_score"] >= 70]

math_passing_total = math_passing["math_score"].count()

math_passing_percent = (math_passing_total / total_students) * 100

math_passing_percent
```

```python
#find percentage of students who passed reading
reading_passing = schools_complete.loc[schools_complete["reading_score"] >= 70]

reading_passing_total = reading_passing["reading_score"].count()

reading_passing_percent = (reading_passing_total / total_students) * 100

percent_passing_reading
```
